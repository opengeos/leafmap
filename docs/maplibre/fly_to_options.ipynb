**Slowly fly to a location**

This source code of this example is adapted from the MapLibre GL JS example - [Slowly fly to a location](https://maplibre.org/maplibre-gl-js/docs/examples/flyto-options).

Uncomment the following line to install [leafmap](https://leafmap.org) if needed.

```python
# %pip install "leafmap[maplibre]"
```

```python
import os
import leafmap.maplibregl as leafmap
```

To run this notebook, you will need an [API key](https://docs.maptiler.com/cloud/api/authentication-key/) from [MapTiler](https://www.maptiler.com/cloud/). Once you have the API key, you can set it as an environment variable in your notebook or script as follows:

```python
# os.environ["MAPTILER_KEY"] = "YOUR_API_KEY"
```

```python
MAPTILER_KEY = os.environ.get("MAPTILER_KEY")
style = f"https://api.maptiler.com/maps/streets/style.json?key={MAPTILER_KEY}"
```

```python
m = leafmap.Map(center=[-74.5, 40], zoom=9, style=style)
m
```

```python
options = {
    "lon": 74.5,
    "lat": 40,
    "zoom": 9,
    "bearing": 0,
    "speed": 0.2,
    "curve": 1,
    "essential": True,
}

m.fly_to(**options)
```
