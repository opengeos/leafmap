**Slowly fly to a location**

```python
import leafmap.maplibregl as leafmap
```

To run this notebook, you will need an [API key](https://docs.maptiler.com/cloud/api/authentication-key/) from [MapTiler](https://www.maptiler.com/cloud/).
```python
m = leafmap.Map(center=[-74.5, 40], zoom=9, style="streets")
m
```

```python
options = {
    "lon": 74.5,
    "lat": 40,
    "zoom": 9,
    "bearing": 0,
    "speed": 0.2,
    "curve": 1,
    "essential": True,
}

m.fly_to(**options)
```
